```python
# На вход данной функции поступает строка с адресом.
def get_street_type(address):
# Создаём список географических пометок exclude_list.
    exclude_list = ['N', 'S', 'W', 'E']
# Метод split() разбивает строку на слова по пробелу.
# В результате получаем список слов в строке и заносим его в переменную address_list.
    address_list = address.split(' ')
# Обрезаем список, оставляя в нём только последний элемент,
# потенциальный подтип улицы, и заносим в переменную street_type.
    street_type = address_list[-1]
# Делаем проверку на то, что полученный подтип является географической пометкой.
# Для этого проверяем его на наличие в списке exclude_list.
    if street_type in exclude_list:
# Если переменная street_type является географической пометкой,
# переопределяем её на второй элемент с конца списка address_list.
        street_type = address_list[-2]
# Возвращаем переменную street_type, в которой хранится подтип улицы.
    return street_type
street_types = melb_df['Address'].apply(get_street_type)
display(street_types)
street_types.value_counts()
```

```python
melb_df['StreetType'] = street_types
# Поубираем дублирующие сокращения которые найдем
street_types = street_types.apply(lambda x: 'Av' if x in ['Ave', 'Avenue'] else x)
street_types = street_types.apply(lambda x: 'Bv' if x in ['Boulevard'] else x)
street_types = street_types.apply(lambda x: 'Pde' if x in ['Parade'] else x)
melb_df['StreetType'] = street_types
# Зададим 10 наиболее частовстречающиеся названия улиц как список_"популярные улицы"
popular_stypes = street_types.value_counts().nlargest(10).index
print(popular_stypes)
# Все остальные названия типов улиц заменим на "other"
melb_df['StreetType'] = street_types.apply(lambda x: x if x in popular_stypes else 'other')
display(melb_df['StreetType'])
melb_df['StreetType'].nunique()
```

```python
# 4.2 Функция которая принимает на вход элемент столбца WeekdaySale и возвращает 1, 
# если день является выходным, и 0 — в противном случае, 
# и создайте столбец Weekend в таблице melb_df с помощью неё.
melb_df['WeekdaySale'].value_counts()

def get_weekend(weekday):
    if weekday == 5 or weekday == 6:
        return 1
    else:
        return 0

weekend_sale = melb_df['WeekdaySale'].apply(get_weekend)
melb_df['Weekend'] = weekend_sale
# Вычислим среднюю цену объекта проданного в выходные (округлим до целых)
round(melb_df['Price'][melb_df['Weekend'] == 1].mean())
```

```python
# 4.4 Приведем данные по опыту работы к числу месяцев
import pandas as pd

def get_experience(arg):
    # запишем переданный аргумент как список слов
    experience_list = arg.split(' ')
    # определим вариации написания слова "год"
    year_list = ['год', 'Год', 'лет', 'Лет', 'года', 'Года']
    # определим вариации написания слова "месяц"
    month_list = ['месяц', 'Месяц', 'месяцев', 'Месяцев', 'месяца', 'Месяца']
    # если последнее слово "год" наврядли нужно учитывать месяцы
    if experience_list[-1] in year_list:
        return int(experience_list[2]) * 12
    # если 4ое слово "месяц" наврядли нужно учитывать года
    elif experience_list[3] in month_list:
        return int(experience_list[2])
    # в остальных случая учтём года и месяцы
    else:
        return (int(experience_list[2]) * 12) + int(experience_list[4])

if __name__ == '__main__':
    experience_col = pd.Series([
        'Опыт работы 8 лет 3 месяца',
        'Опыт работы 3 года 5 месяцев',
        'Опыт работы 1 год 9 месяцев',
        'Опыт работы 3 месяца',
        'Опыт работы 6 лет'
        ])
    experience_month = experience_col.apply(get_experience)
    print(experience_month)
```

```python
# 6.10 Создайте признак длительности поездки trip duration
citibike_df['trip duration'] = citibike_df['stoptime'] - citibike_df['starttime']
# запишем среднюю продолжительность поездки в переменную mean_trip
mean_trip = citibike_df['trip duration'].mean()
# вычислим количество секунд (округлив до целого)
round(mean_trip.total_seconds(),2)
```

```python
# 6.11 Создайте «признак-мигалку» weekend, который равен 1, 
# если поездка начиналась в выходной день (суббота или воскресенье), 
# и 0 — в противном случае. Выясните, сколько поездок начиналось в выходные.

# запишем в переменную день недели
day_of_week = citibike_df['starttime'].dt.dayofweek
# определим функцию, которая вернёт 1 в выходные (5,6) и 0 в будни (0,1,2,3,4)
def is_weekend(arg):
    if arg == 5 or arg ==6:
        return 1
    else:
        return 0
# добавим столбец применив функцию к переменной    
citibike_df['weekend'] = day_of_week.apply(is_weekend)
citibike_df[citibike_df['weekend'] == 1]
```

```python
# 6.12 Создайте признак времени суток поездки time_of_day
time_of_day = citibike_df['starttime'].dt.hour

def day_time(arg):
    if 0 <= arg <= 6:
        return 'night'
    elif 6 < arg <= 12:
        return 'morning'
    elif 12 < arg <= 18:
        return 'day'
    elif 18 < arg <= 24:
        return 'evening'

citibike_df['time of day'] = time_of_day.apply(day_time)
day_trip = citibike_df[citibike_df['time of day']=='day'].count()
night_trip = citibike_df[citibike_df['time of day']=='night'].count()
round(day_trip / night_trip)
```
