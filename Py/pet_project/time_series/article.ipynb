```python
height = [0.6, 0.65, 0.7, 0.72, 0.76, 0.82, 0.85, 0.9, 0.98, 1.01, 1.08, 1.11, 1.24, 1.25]
rand = np.random.uniform(low=0.7, high=1.1, size=(len(height),))
dates = pd.date_range("2020", freq="Q", periods=len(height))
df = pd.DataFrame({'height':height, 'exchange rate':rand}, index=dates)

fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlabel('Дата')
ax.set_ylabel('Наблюдение')
ax.set_title('Наблюдения в диапазоне 05.2020 - 05.2023')
ax.plot(df['height'], color='tab:red', label = 'height')
ax.plot(df['exchange rate'], color='tab:blue', label = 'exchange rate')
ax.legend()
ax.grid()
```
